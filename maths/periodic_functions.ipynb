```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi

%matplotlib inline
```

```python
def f(t):
    if -pi <= t <= 0:
        return -t - pi
    elif 0 <= t <= pi:
        return t - pi
    else:
        raise Exception('undefined')
```

```python
t = np.arange(-pi, pi, 0.1);
ft = [f(_t) for _t in t]
plt.xlabel("t")
plt.ylabel("f(t)")

plt.plot(t - 2*pi, ft, t, ft, t + 2*pi, ft)
```
