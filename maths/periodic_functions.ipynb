```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi

%matplotlib inline
```

```python
def f(t):
    if -pi <= t <= 0:
        return -t - pi
    elif 0 <= t <= pi:
        return t - pi
    else:
        raise Exception('undefined')
```

```python
t = np.arange(-pi, pi, 0.1);
ft = [f(_t) for _t in t]
plt.xlabel("t")
plt.ylabel("f(t)")

plt.plot(t - 2*pi, ft, t, ft, t + 2*pi, ft)
```

```python
# Odd extension of f(t)
def f(t):
    return t*(1 - t)

def f_odd(t, T):
    if 0 < t < T:
        return f(t)
    elif -T < t < 0:
        return -f(-t)
    elif t == 0 or t == T or -t == T:
        return 0
    else:
        raise Exception('undefined ({}, {})'.format(t, T))
        
def f_even(t, T):
    if 0 <= t < T:
        return f(t)
    elif -T <= t < 0:
        return f(-t)
    else:
        raise Exception('undefined ({}, {})'.format(t, T))
```

```python
T = 3
ts1 = np.arange(-T, 0, 0.001);
ft1 = [f_odd(t, T) for t in ts1]

ts2 = np.arange(0, T, 0.001);
ft2 = [f_odd(t, T) for t in ts2]

plt.xlabel("t")
plt.ylabel("f(t)")

plt.plot(ts1, ft1, label="f(t) for -T < t < 0")
plt.plot(ts2, ft2, label="f(t) for 0 < t < T")
plt.legend(loc='best')
```

```python
T = 3
ts1 = np.arange(-T, 0, 0.001);
ft1 = [f_even(t, T) for t in ts1]

ts2 = np.arange(0, T, 0.001);
ft2 = [f_even(t, T) for t in ts2]

plt.xlabel("t")
plt.ylabel("f(t)")

plt.plot(ts1, ft1, label="f(t) for -T <= t < 0")
plt.plot(ts2, ft2, label="f(t) for 0 <= t < T")
plt.legend(loc='best')
```
